## Assignment Statements

In an assignment statement, we make a direction for Python to bind the variable on the **left side of the** = operator to the object produced on the right side. For example, when we write z = x + y, we are telling Python:

 "associate the variable z with the sum of the values associated with the variables x and y."

* **You, Dr. Pi, are riding on a space ship and you asked for an estimated arrival of another ship, operated by your friend Commander Gamma back from intersteller travel. They only sent you the number `7684` due to a glitch in the communication console.
How many days, hours, second later they will arrive?**

**Assume the number 7684 is referring to the number of seconds.**

```python
## declare variables

total_seconds = 7684     #Total number of seconds

## You code goes here

# Calculate days
days = total_seconds // (24 * 3600)  # Number of seconds in a day
remaining_seconds = total_seconds % (24 * 3600) # Remaining seconds using modulo function

# Calculate hours
hours = remaining_seconds // 3600  # Number of seconds in an hour
remaining_seconds = remaining_seconds % 3600

# Calculate minutes
minutes = remaining_seconds // 60  # Number of seconds in a minute
seconds = remaining_seconds % 60

# Print the result
print('Estimated arrival time: {days} days, {hours} hours, {minutes} minutes, {seconds} seconds')
# hmm...

# print the final result
print(f'Estimated arrival time: {days} days, {hours} hours, {minutes} minutes, {seconds} seconds')
```
